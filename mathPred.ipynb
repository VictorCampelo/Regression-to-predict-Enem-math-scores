```python
import numpy as np
from sklearn import linear_model
from sklearn.metrics import explained_variance_score, max_error, mean_absolute_error, r2_score, plot_confusion_matrix

clf = linear_model.LinearRegression()
clf.fit(x_train, y_train)
pred = clf.predict(x_test)
print(clf.score(x_train, y_train))
print(explained_variance_score(y_test, pred))
print(max_error(y_test, pred))
print(mean_absolute_error(y_test, pred))
print(r2_score(y_test, pred))
```

```python
kfolds = KFold(n_splits=10, shuffle=True, random_state=42)

# setup models    
alphas_alt = [14.5, 14.6, 14.7, 14.8, 14.9, 15, 15.1, 15.2, 15.3, 15.4, 15.5]
alphas2 = [5e-05, 0.0001, 0.0002, 0.0003, 0.0004, 0.0005, 0.0006, 0.0007, 0.0008]
e_alphas = [0.0001, 0.0002, 0.0003, 0.0004, 0.0005, 0.0006, 0.0007]
e_l1ratio = [0.8, 0.85, 0.9, 0.95, 0.99, 1]

ridge = make_pipeline(RobustScaler(),
                      linear_model.RidgeCV(alphas=alphas_alt, cv=kfolds))

lasso = make_pipeline(RobustScaler(),
                      linear_model.LassoCV(max_iter=1e7, alphas=alphas2,
                              random_state=42, cv=kfolds))

elasticnet = make_pipeline(RobustScaler(),
                           linear_model.ElasticNetCV(max_iter=1e7, alphas=e_alphas,
                                        cv=kfolds, l1_ratio=e_l1ratio))
                                        
gbr = GradientBoostingRegressor(n_estimators=3000, learning_rate=0.05,
                                   max_depth=4, max_features='sqrt',
                                   min_samples_leaf=15, min_samples_split=10, 
                                   loss='huber', random_state =42)
                                   

lightgbm = lgb.LGBMRegressor(objective='regression', 
                                       num_leaves=4,
                                       learning_rate=0.01, 
                                       n_estimators=5000,
                                       max_bin=200, 
                                       bagging_fraction=0.75,
                                       bagging_freq=5, 
                                       bagging_seed=7,
                                       feature_fraction=0.2,
                                       feature_fraction_seed=7,
                                       verbose=-1,
                                       #min_data_in_leaf=2,
                                       #min_sum_hessian_in_leaf=11
                                       )
                                       

xgboost = xgb.XGBRegressor(learning_rate=0.01, n_estimators=3460,
                                     max_depth=3, min_child_weight=0,
                                     gamma=0, subsample=0.7,
                                     colsample_bytree=0.7,
                                     objective='reg:linear', nthread=-1,
                                     scale_pos_weight=1, seed=27,
                                     reg_alpha=0.00006)
```

```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
import xgboost as xgb
from xgboost.sklearn import XGBRegressor
import datetime
from sklearn.model_selection import GridSearchCV

# Various hyper-parameters to tune
xgb1 = XGBRegressor()
parameters = {'nthread':[4], #when use hyperthread, xgboost may become slower
              'objective':['reg:linear'],
              'learning_rate': [.03, .05, .07], #so called `eta` value
              'max_depth': [3, 4, 5],
              'min_child_weight': [4],
              'silent': [1],
              'colsample_bytree': [0.5, 0.7],
              'n_estimators': [50, 100],
              'reg_lambda':[1e-5, 1e-2, 0.45],
              'subsample':[0.5,0.6,0.7] }

xgb_grid = GridSearchCV(xgb1,
                        parameters, 
                        n_jobs=-1,
                        verbose=True)

xgb_grid.fit(x_train,y_train)

print(xgb_grid.best_score_)
print(xgb_grid.best_params_)
```

```python
best_xgb_model = XGBRegressor(
    colsample_bytree= 0.7, learning_rate= 0.05, 
    max_depth=4, min_child_weight= 4, n_estimators= 100, 
    nthread=-1, objective= 'reg:linear', 
    reg_lambda= 0.45, silent=1, subsample= 0.5)
best_xgb_model.fit(x_train,y_train)

pred = best_xgb_model.predict(x_test)
print("Score: "+ str(best_xgb_model.score(x_train, y_train)))

print(max_error(y_test, pred))
print(mean_absolute_error(y_test, pred))
print(r2_score(y_test, pred))
```

```python
from sklearn.feature_selection import SelectKBest, SelectPercentile, f_regression, f_classif
from sklearn.metrics import mean_absolute_error
from sklearn.ensemble import RandomForestRegressor
import warnings
warnings.filterwarnings("ignore")

k_vs_score = []

for k in range(2,42,2):
  selector = SelectKBest(score_func=f_regression, k=k)

  Xtrain2 = selector.fit_transform(x_train, y_train)
  Xval2 = selector.transform(x_test)

  mdl = XGBRegressor(
    colsample_bytree= 0.7, learning_rate= 0.05, 
    max_depth=4, min_child_weight= 4, n_estimators= 100, 
    nthread=-1, objective= 'reg:linear', 
    reg_lambda= 0.45, silent=1, subsample= 0.5)
  
  best_xgb_model.fit(x_train,y_train)

  mdl.fit(Xtrain2, y_train)

  p = mdl.predict(Xval2)

  score = mean_absolute_error(y_test, p)
  print("k = {} - MAE = {}".format(k, score))

  k_vs_score.append(score)
```

```python
pd.Series(k_vs_score, index=range(2,42,2)).plot(figsize=(10,7))
```
